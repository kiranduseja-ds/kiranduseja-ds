# Model Selection:
1) Using The Sentiment Intensity Analyzer (VADER).
2) Using The Confusion Matrix Classification Report to get Model Accuracy.

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
se=SentimentIntensityAnalyzer()
```

```python
s='The movie was worst MOVIE I have ever seen'
```

```python
se.polarity_scores(s)
```

```python
s1='wow I am very Happy with the product,loved it '
```

```python
se.polarity_scores(s1)
```

```python
corpus['Polarity_score']=corpus['review'].apply(lambda x:se.polarity_scores(x))
```

```python
corpus
```

```python
corpus['compound']=corpus['Polarity_score'].apply(lambda x:x['compound'])
```

```python
corpus
```

```python
corpus.iloc[10]['review']
```

```python
age=45
print('You are eligible to vote') if(age>=18) else print('You are not eligible to vote')
```

```python
corpus['sentiment']=corpus['compound'].apply(lambda x:'pos' if x>=0 else 'neg')
```

```python
corpus
```

```python
corpus['label'].value_counts()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
y_true=corpus['label']
y_pred=corpus['sentiment']
```

```python
confusion_matrix(y_true,y_pred)
```

```python
print(classification_report(y_true,y_pred))
```

# CONCLUSION:
This project analyzed Amazon product reviews using the Sentiment Intensity Analyzer to classify sentiments as positive or negative.
