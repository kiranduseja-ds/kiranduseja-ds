```python
# converting y to One Hot Encoding so 1 column for each degit.

y_cat_train=to_categorical(y_train)

y_cat_test=to_categorical(y_test)
```

```python
x_train=x_train.reshape(60000,28,28,1) # value 3 is for colourfull image
```

```python
x_test=x_test.reshape(10000,28,28,1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Flatten
```

```python
model=Sequential()
```

```python
model.add(Conv2D(filters=32,kernel_size=(4,4),input_shape=(28,28,1),activation='relu'))
```

```python
model.add(MaxPool2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(filters=16,kernel_size=(3,3),activation='relu')) # imput_shape will come only first CNN Layer
model.add(MaxPool2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add((Dense(128,activation='relu'))) # ANN Hidden Layer
model.add((Dense(64,activation='relu')))

model.add((Dense(10,activation='softmax'))) # ANN Output Layer

model.compile(loss='categorical_crossentropy')
```

```python
model.fit(x_train,y_cat_train,epochs=50)
```
